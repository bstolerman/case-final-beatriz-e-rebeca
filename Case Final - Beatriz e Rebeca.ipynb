```python
import pandas as pd
import altair as alt
```

```python
dados_seguradoras = pd.read_csv("insurance2.csv")
dados_seguradoras
```

1) Qual Ã© o nÃºmero total de fumantes?

```python
pessoas_fumantes = dados_seguradoras[['age', 'smoker']].groupby('age', as_index=False).count()
```

```python
pessoas_fumantes
```

```python
alt.Chart(dados_seguradoras, title="Quantidade de segurados que sÃ£o fumantes").mark_bar().encode(
    x= alt.X('age', title='Segurados por idade'),
    y= alt.Y('smoker', title='Fumantes'),
    color = alt.Color('age')
).interactive()
```

2) Qual regiÃ£o aciona mais o seguro?

```python
regiao_que_mais_aciona = dados_seguradoras[['insuranceclaim', 'region']].groupby('region', as_index=False).count()
```

```python
regiao_que_mais_aciona
```

```python
alt.Chart(regiao_que_mais_aciona, title="RegiÃ£o que mais aciona o seguro").mark_bar().encode(
    x= alt.X('region', title='RegiÃ£o'),
    y= alt.Y('insuranceclaim', title='Acionamento do seguro'),
    color = alt.Color('region')
).interactive()
```
